```python
symbol = 'AAPL'
api_url = f"https://cloud.iexapis.com/stable/stock/{symbol}/quote?token={IEX_CLOUD_API_TOKEN}"   #created f string that replaces symbol
#print(api_url)
data = requests.get(api_url).json()   #status == 200 means OK created a json file
# now the data can be used as a dictionary
data
```

```python
price = data['latestPrice']
market_cap = data['marketCap']
name = data['companyName']
```

```python
my_columns = ['symbol', 'name', 'price' , 'market cap' , 'shares to buy']
final_dataframe = pd.DataFrame(columns = my_columns)
```

```python
final_dataframe.append(
    pd.Series(
        [
            symbol,
            name,
            price,
            market_cap,
            'N/A'
        ],
        index = my_columns 
    ),
    ignore_index=True  #it is must dont know why 
)
```

```python
for stock in stocks['Ticker']:
    api_url = f"https://cloud.iexapis.com/stable/stock/{stock}/quote?token={IEX_CLOUD_API_TOKEN}"
    data = requests.get(api_url).json()
    price = data['latestPrice']
    name = data['companyName']
    market_cap = data['marketCap']
    
    final_dataframe = final_dataframe.append(
        pd.Series(
            [
                stock,
                name,
                price,
                market_cap,
                'N/A'
            ],
            index = my_columns 
        ),
        ignore_index=True  #it is must dont know why 
    )
    
```

```python
amount = float(input("enter amount you want to invest:"))
amount_per_share = amount / len(final_dataframe)
amount_spend=0
```

```python
size = len(final_dataframe)
for i in range(size):
    price = final_dataframe.loc[i,'price']
    final_dataframe.loc[i,'shares to buy'] = math.floor(amount_per_share/price)
    amount_spend = amount_spend + final_dataframe.loc[i,'shares to buy'] * price
```
